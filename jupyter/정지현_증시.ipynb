# 인기 검색 목록

```python
# import requests
# from bs4 import BeautifulSoup

def popular():
    req = requests.get('https://finance.naver.com/sise/') # requests.get으로 url의 html 정보들을 가져옴
    html = req.text # html 정보에서 text 정보만 빼서 html 변수에 넣어줌
    soup = BeautifulSoup(html, 'html.parser') # html.parser 로 파싱
    
    my_stock = soup.select( #  my_stock 변수에 select 할 정보를 찾아서 넣어줌
    '.lst_pop'
    )
    
    for stock_name in my_stock: 
        print(stock_name.text)

popular()
```

# 국내증시 > 시가총액 > 코스피

```python
# import requests
# from bs4 import BeautifulSoup

def return_value(address):
    res = requests.get(address)
    soup = BeautifulSoup(res.content, 'html.parser')

    section = soup.find('tbody')
    items = section.find_all('tr', onmouseover="mouseOver(this)")
    for item in items:
        basic_info = item.get_text()
        sinfo = basic_info.split("\n")
        print("\t" + sinfo[1] + "\t\t" + sinfo[2] + "\t\t\t" + sinfo[3])


baseaddress = 'https://finance.naver.com/sise/sise_market_sum.naver?&page='
for i in range(1,35):
    return_value(baseaddress+str(i))
```

# 특정 종목 일별 시세

```python
# import requests
# import pandas as pd
# import xmltodict
# import json

def mystock_his():
    stockCode=input('검색할 종목의 코드를 입력하세요.')
    count='2500'
    url = f'https://fchart.stock.naver.com/sise.nhn?symbol={stockCode}&timeframe=day&count={count}&requestType=0'

    rs = requests.get(url)
    dt = xmltodict.parse(rs.text)
    js = json.dumps(dt,indent=4)
    js = json.loads(js)

    data = pd.json_normalize(js['protocol']['chartdata']['item'])
    df = data['@data'].str.split('|',expand=True)
    df.columns = ['날짜','시가','고가','저가','종가','거래량']
    print(df)

mystock_his()
```
